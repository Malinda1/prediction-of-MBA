```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

# Connecting to the js library
init_notebook_mode(connected  = True)

# Initialize offline plot
cf.go_offline()
```

```python
# Importing dataset and create a copy of dataset

data_01 = pd.read_csv(f'mba_decision_dataset.csv')

data_v1 = data_01.copy()

data_v1.head()
```

```python
# Gender Distribution

plt.figure(figsize = (10, 6))

count_of_gender = data_v1['Gender'].value_counts().reset_index()
count_of_gender.columns = ['Gender', 'Count']
sns.barplot(
    x = 'Gender',
    y = 'Count',
    data = count_of_gender,
    hue = 'Gender',
    palette = 'Set2'
)
plt.title('Gender Distribution')
plt.xlabel('Gneder')
plt.ylabel('Age')
plt.show()
```

```python
# Undergraduate Major Distribution

plt.figure(figsize = (10, 6))

count_of_Undergraduate_Major = data_v1['Undergraduate Major'].value_counts().reset_index()
count_of_Undergraduate_Major.columns = ['Undergraduate Major', 'Count']

sns.barplot(
    x = 'Undergraduate Major',
    y = 'Count',
    data = count_of_Undergraduate_Major,
    hue = 'Undergraduate Major',
    palette = 'Set2'
)

plt.title('Distribution of Undergraduate Major')
plt.xlabel('Undergraduate Major')
plt.ylabel('Count of the Undergraduate Major')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(
    x = 'Current Job Title',
    y = 'Annual Salary (Before MBA)',
    data = data_v1,
    hue = 'Current Job Title',
    palette = 'Set2'
)

plt.title('Annual salary of each job category')
plt.xlabel('Cuurent job Title')
plt.ylabel('Annual Salary')
plt.show()
```
